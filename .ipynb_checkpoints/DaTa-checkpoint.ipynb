```python
mean_fd= new.fixed_acidity.mean()
mean_vd= new.volatile_acidity.mean()
mean_cd= new.citric_acid.mean()
mean_rs= new.residual_sugar.mean()
mean_cl= new.chlorides.mean()
mean_fsd= new.free_sulfur_dioxide.mean()
mean_tsd= new.total_sulfur_dioxide.mean()
mean_dy= new.density.mean()
mean_pH= new.pH.mean()
mean_sl= new.sulphates.mean()
mean_al= new.alcohol.mean()
mean_ql= new.quality.mean()
```

```python
print(f'fixed_acidity:{mean_fd}')
print(f'volatile_acidity:{mean_vd}')
print(f'citric_acid:{mean_cd}')
print(f'residual_sugar:{mean_rs}')
print(f'chlorides:{mean_cl}')
print(f'free_sulfur_dioxide:{mean_fsd}')
print(f'total_sulfur_dioxide:{mean_tsd}')
print(f'density:{mean_dy}')
print(f'pH:{mean_pH}')
print(f'sulphates:{mean_sl}')
print(f'alcohol:{mean_al}')
print(f'quality:{mean_ql}')
```
